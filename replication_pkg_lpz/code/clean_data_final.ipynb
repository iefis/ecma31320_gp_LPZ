#### Identify individuals who have been 6-18 years in the sample

```python
# step 1
# list of years of observations
list_yr = [v[0] for k, v in varDict.items()]
list_yr1 = list(set(list_yr) & (set(range(1969,2012))))
df_sample = pd.DataFrame()
set_sample = set()
for yr in list_yr1:
    var_age = inv_varDict[(yr,"age")]
    var_race = inv_varDict[(yr,"race")]
    var_hstat = inv_varDict[(yr,"hstat")]
    var_hval = inv_varDict[(yr,"hval")]
    var_yage = inv_varDict[(yr,"yage")]
    var_hhnum = inv_varDict[(yr,"hhnum")]
    var_nseq = inv_varDict[(yr,"nseq")]
    var_relation = inv_varDict[(yr,"relation")]
    var_list = [var_age, var_race, 
                var_hstat,  var_hval, var_yage, "sex",
                var_hhnum, var_nseq, var_relation]
    name_list = {var_age:"age", var_race:"race",
                var_hstat:"hstat", var_hval:"hval",
                var_yage:"yage", var_hhnum:"hhnum", 
                var_nseq:"nseq", var_relation:"relation"}
    age_ = ((df1[var_age] >= 6) & (df1[var_age] <= 18))
    relation_ = ((df1[var_relation] == 3) | (df1[var_relation] == 30) | 
                 (df1[var_relation] == 33) | (df1[var_relation] == 35))
    set_add = set(df1.ID[age_ & relation_]).difference(set_sample)
    set_sample = set_add | set_sample
    df_sample1 = df1[var_list][df1.ID.isin(set_add)]
    df_sample1 = df_sample1.rename(columns=name_list)
    df_sample1["year"] = yr
    df_sample = pd.concat([df_sample, df_sample1],axis=0)
df_sample["ID"] = df_sample.index.values
df_sample = df_sample.reset_index()
df_sample = df_sample.drop(columns="index")
```

```python
# search year by year from 1969 to 2019-(25-18) = 2012
# list_yr1 = list(set(list_yr) & (set(range(1969,2000))))
df_mov_age = pd.DataFrame()
set_mov = set()
for yr in list_yr1:
    var_mov = inv_varDict[(yr,"mov")]
    var_age = inv_varDict[(yr,"age")]
    var_race = inv_varDict[(yr,"race")]
    var_reason = inv_varDict[(yr,"reason")]
    var_hstat = inv_varDict[(yr,"hstat")]
    var_hval = inv_varDict[(yr,"hval")]
    var_yage = inv_varDict[(yr,"yage")]
    var_hhnum = inv_varDict[(yr,"hhnum")]
    var_nseq = inv_varDict[(yr,"nseq")]
    var_relation = inv_varDict[(yr,"relation")]
    move_ = (df1[var_mov] == 1)
    age_ = ((df1[var_age] >= 6) & (df1[var_age] <= 18))
    set_ind1 = set(df1.ID[move_ & age_]).difference(set_mov)
    set_mov = set_ind1 | set_mov
    var_list = [var_age, var_race, var_mov, var_reason,
                var_hval, var_yage, var_hstat,  
                var_hhnum, var_nseq, var_relation]
                # var_nstate, var_nreg, var_nroom, var_pproom]
    name_list = {var_age:"mov_age", var_race:"mov_race", var_mov:"mov", 
                var_reason:"reason", var_hval:"mov_hval",
                var_yage:"mov_yage", var_hhnum:"mov_hhnum", var_hstat:"mov_hstat",
                var_nseq:"mov_nseq", var_relation:"mov_relation"}
    df_mov_age1 = df1[var_list][df1.ID.isin(set_ind1)]
    df_mov_age1 = df_mov_age1.rename(columns=name_list)
    df_mov_age1["mov_year"] = yr
    df_mov_age = pd.concat([df_mov_age, df_mov_age1],axis=0)
df_mov_age["ID"] = df_mov_age.index.values
```
